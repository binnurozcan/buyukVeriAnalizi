```
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords

turkish_stopwords = stopwords.words('turkish')
vectorizer = CountVectorizer(ngram_range=(2,2), stop_words=turkish_stopwords)

grouped = subat11.groupby('şehir')
for sehir, df in grouped:
    tweets = df['tweet']
    X = vectorizer.fit_transform(tweets)
    ngram_counts = X.sum(axis=0)
    ngram_freqs = [(word, ngram_counts[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    sorted_ngrams = sorted(ngram_freqs, key=lambda x: x[1], reverse=True)
    top_ngrams = sorted_ngrams[:5]
    print(sehir, top_ngrams)
    
```

```
import numpy as np

with open("/gdrive/MyDrive/depremTweetler/sehirler.txt", "r") as f:
    icerik = f.read()
    kelimeler = icerik.split()
    sehirler = np.array(kelimeler)

print(sehirler)
```
